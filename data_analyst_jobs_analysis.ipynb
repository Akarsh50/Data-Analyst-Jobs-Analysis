```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=['title', 'company_name', 'location'])
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.loc[:, 'via'] = df['via'].fillna('Unknown')
```

```python
df.loc[:, 'schedule_type'] = df['schedule_type'].fillna('Not Specified')
```

```python
df = df.drop(columns=['index'])
```

```python
df['title'] = df['title'].str.lower().str.strip()
df['company_name'] = df['company_name'].str.strip()
df['location'] = df['location'].str.strip()
df['work_from_home']=df['work_from_home'].str.strip()
```

```python
import ast
def safe_parse_tokens(x):
    # Case 1: Already a list
    if isinstance(x, list):
        return x
    
    # Case 2: NaN or empty
    if pd.isna(x) or x == '':
        return []
    
    # Case 3: String that looks like a list
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except (ValueError, SyntaxError):
            return []  # fallback for malformed rows
    
    return []
```

```python
df.loc[:, 'description_tokens'] = df['description_tokens'].apply(safe_parse_tokens)
```

```python
df['description_tokens'].head(10)
df['description_tokens'].apply(type).value_counts()
```

```python
skill_map = {
    'power_bi': 'power bi',
    'powerbi': 'power bi',
    'ms excel': 'excel',
    'mysql': 'sql',
    'mssql': 'sql',
    'postgresql': 'sql',
    'postgres': 'sql',
    'redshift': 'aws',
    'snowflake': 'snowflake',
    'sas': 'sas',
    'spss': 'spss',
    'jupyter': 'python',
    'matplotlib': 'python',
    'pandas': 'python'
}
```

```python
def normalize_skills(skill_list):
    normalized = []
    for skill in skill_list:
        skill = skill.lower()
        normalized.append(skill_map.get(skill, skill))
    return list(set(normalized))  # remove duplicates

df.loc[:, 'description_tokens'] = df['description_tokens'].apply(normalize_skills)
```

```python
df=df[['title','company_name','location','via','schedule_type','work_from_home','description_tokens']]
```

```python
df.loc[:, 'work_from_home'] = df['work_from_home'].fillna(False)
```

```python
import json

df['description_tokens'] = df['description_tokens'].apply(
    lambda x: json.dumps(x) if isinstance(x, list) else json.dumps([])
)
```
